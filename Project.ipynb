```python
# if hours of absentisim is less than 7, then its fine,otherwise it is not acceptable
```

```python
df1['Absenteeism time in hours'] = df1['Absenteeism time in hours'].apply(lambda x:0 if x >=7 else 1)
df1['Absenteeism time in hours'].value_counts()
```

```python
df1['Absenteeism time in hours'].value_counts()/len(df1['Absenteeism time in hours'])*100
```

```python
#Scaling the data by using Standard Scaler
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
df_scaled = sc.fit_transform(df1.drop(columns = ['Absenteeism time in hours','ID']))
df_scaled
```

```python
df_scaled = pd.DataFrame(df_scaled, columns = (df1.drop(columns = ['Absenteeism time in hours','ID']).columns))
df_scaled
```

```python
x = df_scaled
y = df1['Absenteeism time in hours']
```

```python
x.head()
```

```python
y.head()
```

```python
#Model training
```

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest = train_test_split(x,y,test_size = 0.25)
```

```python
print(xTrain.shape)
print(xTest.shape)
print(yTrain.shape)
print(yTest.shape)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Build clustering and classification
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(xTrain,yTrain)
predict1 = lr.predict(xTest)
acc1 = accuracy_score(predict1,yTest)

#DecisionTree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(xTrain,yTrain)
predict2 = dt.predict(xTest)
acc2 = accuracy_score(predict2,yTest)

#RandomForest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(xTrain,yTrain)
predict3 = rf.predict(xTest)
acc3 = accuracy_score(predict3,yTest)

#SVM
from sklearn.svm import SVC
svc = SVC()
svc.fit(xTrain,yTrain)
predict4 = svc.predict(xTest)
acc4 = accuracy_score(predict4,yTest)

print('Logistic Regression: ',acc1)
print('Decision Tree: ',acc2)
print('Random Forest: ',acc3)
print('SVM: ',acc4)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
vc = VotingClassifier(estimators = [('lr',lr),('dt',dt),('rf',rf),('svm',svc)],flatten_transform=True,verbose = True)
```

```python
vc.fit(xTrain,yTrain)
pred5 = vc.predict(xTest)
acc5 = accuracy_score(yTest,pred5)
```

```python
pre1_lr = lr.predict(xTrain)
train_acclr = accuracy_score(pre1_lr,yTrain)

pre1_rf = rf.predict(xTrain)
train_accrf = accuracy_score(pre1_rf,yTrain)
pre1_svc = svc.predict(xTrain)
train_accsvc = accuracy_score(pre1_svc,yTrain)
pre1_vc = vc.predict(xTrain)
train_accvc = accuracy_score(pre1_vc,yTrain)
```
